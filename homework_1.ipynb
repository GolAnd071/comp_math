# Problem 1

```python
def join_two_matrices(A, B):
    return np.concatenate((A, B), axis=1)

def construct_array(M, a, b):
    return M[a, b]

def trace_and_nonzero_diag(A):
    return (np.trace(A), np.count_nonzero(np.diagonal(A)))

def change_negative_zero(A, B):
    C = B.copy()
    C[A < 0] = 0
    return C

def scalar_product_of_columns(A, B):
    return A.T @ B
```

```python
A, B = np.ones((5, 2)), np.eye(5)
join_two_matrices(A, B)
```

```python
#     
mean = 0
std = 50

#  
N = 1000

#   
M = 50

#     
eig_vals = []

#  M      
for i in range(M):
    #    N x N
    mat = np.random.normal(mean, std, size=(N, N))
    #   
    mat = mat + mat.T
    #   
    eigs = np.linalg.eigvals(mat)
    #    
    eig_vals.extend(eigs)

#    
num_bins = 100

#    
plt.hist(eig_vals, bins=num_bins)

#         
bin_means, bin_edges, _ = plt.hist(eig_vals, bins=num_bins)
bin_centers = 0.5 * (bin_edges[1:] + bin_edges[:-1])
bin_stddevs = []
for i in range(num_bins):
    indices = np.where((eig_vals >= bin_edges[i]) & (eig_vals <= bin_edges[i+1]))
    bin_stddevs.append(np.std([eig_vals[i] for i in indices[0]]))

#   
plt.errorbar(bin_centers, bin_means, yerr=bin_stddevs, fmt='kx', ecolor='r', elinewidth=2, capsize=4)

plt.title('Histogram of Eigenvalues')
plt.xlabel('Eigenvalue')
plt.ylabel('Frequency')
plt.show()
```

```python
#     
mean = 0
std = 50

#  
N = 1000

#   
M = 50

#       
rho_vals = []

#  M        
for i in range(M):
    #    N x N
    mat = np.random.normal(mean, std, size=(N, N))
    #   
    mat = mat + mat.T
    #   
    eigs = np.linalg.eigvals(mat)
    #  
    eigs.sort()
    #   
    rhos = [eigs[i + 1] - eigs[i] for i in range(len(eigs) - 1)]
    #    
    rho_vals.extend(rhos)

#    
num_bins = 200

#   
plt.hist(rho_vals, bins=num_bins)

plt.title('Histogram of Distances Between Closest Eigenvalues')
plt.xlabel('Distance')
plt.ylabel('Frequency')
plt.show()
```

```python
xs = np.linspace(0.1, 100, 1000)

lin_ys = xs

plt.xscale('log')
plt.yscale('log')

plt.plot(xs, lin_ys * 1000)

plt.hist(rho_vals, bins=num_bins)

plt.show()
```

 $\tilde{f}_{c_1 c_2}(\rho) = |c_1| \rho e^{-|c_2| \rho^2}$.

```python
fig, (ax1, ax2) = plt.subplots(1, 2)

#     
mean = 0
std = 50

#  
N = 1000

#   
M = 50

#     
eig_vals = []

#       
rho_vals = []

#  M      
for i in range(M):
    #    N x N
    mat = np.asmatrix(np.random.normal(mean, std, size=(N, N)) + 1j * np.random.normal(mean, std, size=(N, N)))
    #   
    mat = mat + mat.H
    #   
    eigs = np.linalg.eigvals(mat)
    #  
    eigs.sort()
    #   
    rhos = [eigs[i + 1] - eigs[i] for i in range(len(eigs) - 1)]
    #    
    eig_vals.extend(eigs)
    rho_vals.extend(rhos)

#    
num_bins = 100

#    
ax1.hist(np.asarray(eig_vals).real, bins=num_bins)

#         
bin_means, bin_edges, _ = ax1.hist(np.asarray(eig_vals).real, bins=num_bins)
bin_centers = 0.5 * (bin_edges[1:] + bin_edges[:-1])
bin_stddevs = []
for i in range(num_bins):
    indices = np.where((eig_vals >= bin_edges[i]) & (eig_vals <= bin_edges[i+1]))
    bin_stddevs.append(np.std([eig_vals[i] for i in indices[0]]))

#   
ax1.errorbar(bin_centers, bin_means, yerr=bin_stddevs, fmt='kx', ecolor='r', elinewidth=2, capsize=4)

ax1.set_title('Histogram of Eigenvalues')
ax1.set_xlabel('Eigenvalue')
ax1.set_ylabel('Frequency')

#   
ax2.hist(np.asarray(rho_vals).real, bins=num_bins)

ax2.set_title('Histogram of Distances')
ax2.set_xlabel('Distance')
ax2.set_ylabel('Frequency')

plt.show()
```

```python
xs = np.linspace(0.1, 100, 1000)

sqr_ys = xs ** 2

plt.xscale('log')
plt.yscale('log')

plt.plot(xs, sqr_ys * 140)

plt.hist(rho_vals, bins=num_bins)

plt.show()
```
